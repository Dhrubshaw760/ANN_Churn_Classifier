```python
import tensorflow as tf
import pickle
import pandas as pd
import numpy as np
```

```python
### Load the ANN trained model, scaler pickle, onehot pickle file

from tensorflow.keras.models import load_model
model = load_model('model.h5')

with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
with open('onehot_encoder_geo.pkl', 'rb') as f:
    onehot_encoder_geo = pickle.load(f)
with open('label_encoder_gender.pkl', 'rb') as f:
    label_encoder_gender = pickle.load(f)
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000,
}
```

```python
# Encode Gender (Categorical variable)
input_data['Gender'] = label_encoder_gender.transform([input_data['Gender']])
```

```python
# Encode Geography
ohe = onehot_encoder_geo.transform([[input_data['Geography']]]).toarray()
ohdf = pd.DataFrame(ohe, columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
ohdf
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
# Dict to DataFrame :- (Just make the dictionary a list)
input_df = pd.DataFrame([input_data])

input_df.drop(['Geography'], axis=1, inplace=True)
```

```python
# Combine one-hot encoded columns with input data
df = pd.concat([input_df, ohdf], axis=1)
df.columns.values.shape
```

```python
# Scaling of features
scaled_data = scaler.transform(df)
scaled_data
```

```python
pred = model.predict(scaled_data)
```
